```python
#walmart_ca-ecommerce_product_details
import pandas as pd
import numpy as np
```

```python
#import the dataset
product_details=pd.read_csv('walmart_ca-ecommerce_product_details.csv')
product_details.head(2)
```

```python
#get missing columns data
product_details.info()
```

```python
#Columns to drop
# 1. Package Size,Postal Code : these are null value columns so these should be dropped
# 2. Uniq Id,Product Url,Gtin : These are unneccessary columns in our data set thus these should be dropped too

# index location for the columns to be dropped
#Package Size[10],Postal Code[12],Uniq Id[0],Product Url[2],Gtin[9]

#Drop multiple columns by names
product_details.drop(['Package Size','Postal Code','Uniq Id','Product Url','Gtin'], axis=1, inplace=True)
```

```python
#check if Columns are deleted
product_details.head(2)
```

```python
#Handle missing values in columns
#get missing columns data
product_details.info()
```

```python
#Handle missing Data Columns
# from the above info , Brand is the only column with missing data
# Brand is object and its catergorical data there fore we can use mode()[]
#product_details.Brand.mode()[0]
product_details['Brand'].fillna(product_details.Brand.mode()[0],inplace=True)
```

```python
# Check whether missing columns have been sorted
product_details.info()
```

# Find out whether there is misspelled Data which can result to duplicates
  Major coumns are 
         1 ----------Brand

```python
#Get Product Names
#product_details['Product Name'].unique()
data_frame_for_Product_name=pd.DataFrame(product_details['Product Name'].unique(),columns=['Product_Name'])
data_frame_for_Product_name
```

```python
#count Total Number of Product Names
data_frame_for_Product_name.Product_Name.count()
```

```python
#Get Brands
Dataframe_Brand=pd.DataFrame(product_details['Brand'].unique(), columns=['Brand'])
Dataframe_Brand
# from the code above, You can still Clean further in the Brand Column
#  ----Ray-Ban can be converted to Ray Ban , its a mistype
```

```python
#More on Brand Column
# ----Ray-Ban can be converted to Ray Ban , its a mistype
product_details['Brand'].replace(['Ray-Ban'],['Ray Ban'],inplace=True)
```

```python
# confirm whether ---Ray-Ban has been converted to Ray Ban
Dataframe_Brand
```
