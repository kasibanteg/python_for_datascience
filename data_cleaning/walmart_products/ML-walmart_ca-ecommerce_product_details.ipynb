# Sorting and Further cleaning data for ML
#Identify Posssible Features and Labels first



```python
#Features ---1. Product Name (string--object)--to be converted to Numeric
#            2. Item Number  (string--object)--to be converted to Numeric
#            3. Brand        (string--object)--to be converted to Numeric
#            4. List Price   (float)
#            5. Sale Price   (float)

# Label   ---Available

#NB : All fetaures and Labels should be Numeric BUT for this case,ONLY features will be converted to Numerics 
#through Label Econding Then OHE to remove mathematical weightage since original columns are under catergorical Data
```

```python
# Get the ML Dataset to be Used basing on Features and Labels
ML_dataset=dataset.iloc[:,[1,5,6,3,4,8]]
ML_dataset


```

```python
# Convert Cartegorical Features to Numerics
#Features ---1. Product Name (string--object)--to be converted to Numeric
#            2. Item Number  (string--object)--to be converted to Numeric
#            3. Brand        (string--object)--to be converted to Numeric
```

```python
#Am Eliminating Product Name since the Name has LONG Characters am going to cosinder(BRAND, ITEM CODE,LIST PRICE and SALE PRICE)

#BrandDummies=pd.get_dummies(dataset['Brand'])
#BrandDummies
# Brand and Item Number
Two_features_get_dummies=(pd.concat([pd.get_dummies(dataset['Brand']),pd.get_dummies(dataset['Item Number'])], axis=1))
Two_features_get_dummies.head(3)
```

```python
ForMLdata=pd.concat([Two_features_get_dummies,ML_dataset.iloc[:,[3,4,5]]], axis=1)
ForMLdata.head(4)
```

```python
#If you want, also convert AVAILABLE labels to numeric.
#see below
# I wont make it permanet for now


#dealling with Available.
# its odinal data
#True=1,False=0
# replace values
ForMLdata.Available.replace(['True','False'],[1,0],inplace=False)


```
